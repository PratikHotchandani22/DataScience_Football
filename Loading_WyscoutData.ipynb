### Reading competitions file

### Reading Matches data

```python
england_file = "/Users/pratikhotchandani/Downloads/Github/DataScience_Football/WyscoutData/matches/matches_England.json"
france_file = "/Users/pratikhotchandani/Downloads/Github/DataScience_Football/WyscoutData/matches/matches_France.json"
germany_file = "/Users/pratikhotchandani/Downloads/Github/DataScience_Football/WyscoutData/matches/matches_Germany.json"
spain_file = "/Users/pratikhotchandani/Downloads/Github/DataScience_Football/WyscoutData/matches/matches_Spain.json"
italy_file = "/Users/pratikhotchandani/Downloads/Github/DataScience_Football/WyscoutData/matches/matches_Italy.json"
worldcup_file = "/Users/pratikhotchandani/Downloads/Github/DataScience_Football/WyscoutData/matches/matches_World_Cup.json"
uefa_file = "/Users/pratikhotchandani/Downloads/Github/DataScience_Football/WyscoutData/matches/matches_European_Championship.json"

with open(england_file) as f:
    data = json.load(f)
    df_england_matches = pd.DataFrame(data)
```

```python
df2 = df_england_matches[['date']]
```

```python
df2['date_column'] = df2['date'].apply(lambda x: parser.parse(x).strftime("%d-%m-%Y"))
```

```python
# Convert 'date_column' to datetime type
df2['date_column'] = pd.to_datetime(df2['date_column'], format='%d-%m-%Y')

# Extract unique years
unique_years = df2['date_column'].dt.year.unique()

# Print the unique years
print(unique_years)
```

```python
with open(uefa_file) as f:
    data = json.load(f)
    df_france_matches = pd.DataFrame(data)
```

```python
df = df_france_matches[['date']]
df['date_column'] = df['date'].apply(lambda x: parser.parse(x).strftime("%d-%m-%Y"))
# Convert 'date_column' to datetime type
df['date_column'] = pd.to_datetime(df['date_column'], format='%d-%m-%Y')

# Extract unique years
unique_years = df['date_column'].dt.year.unique()

# Print the unique years
print(unique_years)
```
