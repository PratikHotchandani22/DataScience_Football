In this tutorial we demonstrate how to check equal size of two samples using the sign test. To do so, we provide an example in which we check if Heung-Min Son shoots with both feet the same number of times.



```python
import pandas as pd
import numpy as np
import json
# plotting
import matplotlib.pyplot as plt
#opening data
import os
import pathlib
import warnings

pd.options.mode.chained_assignment = None
warnings.filterwarnings('ignore')
```

Loading players data

```python
folder_path = '/Users/pratikhotchandani/Downloads/Github/DataScience_Football/WyscoutData'
# Specify the name of the CSV file
file_name = 'players.json'

# Combine the folder path and file name to create the full file path
file_path = os.path.join(folder_path, file_name)

# Check if the file exists
if os.path.exists(file_path):
    # Open the CSV file
    with open(file_path, 'r') as file:
        json_data = json.load(file)
        df_players = pd.DataFrame(json_data)
else:
    print("File does not exist!")

df_players
```

We would also need data for events, for this we will consider only events_england file

```python
england_event_file = "/Users/pratikhotchandani/Downloads/Github/DataScience_Football/WyscoutData/events/events_England.json"
with open(england_event_file) as f:
    data = json.load(f)
    df_england_events = pd.DataFrame(data)

df_england_events
```

### Preparing dataset

```python
#take shots only
shots = df_england_events.loc[df_england_events['subEventName'] == 'Shot']
#look for son's id
son_id = df_players.loc[df_players["shortName"] == "Son Heung-Min"]["wyId"].iloc[0]
#get son's shot
son_shots = shots.loc[shots["playerId"] == son_id]

#left leg shots
lefty_shots = son_shots.loc[son_shots.apply (lambda x:{'id':401} in x.tags, axis = 1)]
#right leg shots
righty_shots = son_shots.loc[son_shots.apply (lambda x:{'id':402} in x.tags, axis = 1)]

#create list with ones for left foot shots and -1 for right foot shots
l = [1] * len(lefty_shots)
l.extend([-1] * len(righty_shots))
```

```python
lefty_shots
```

```python
righty_shots
```

```python
from statsmodels.stats.descriptivestats import sign_test
test = sign_test(l, mu0 = 0)
pvalue = test[1]

if pvalue < 0.05:
    print("P-value amounts to", str(pvalue)[:5], "- We reject null hypothesis - Heung-Min Son is not ambidextrous")
else:
    print("P-value amounts to", str(pvalue)[:5], " - We do not reject null hypothesis - Heung-Min Son is ambidextrous")
```
