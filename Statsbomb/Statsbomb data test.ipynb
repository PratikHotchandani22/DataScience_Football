```python
import pandas as pd
import numpy as np
```

```python
from mplsoccer import Sbopen

# instantiate a parser object
parser = Sbopen()
```

```python
df_competition = parser.competition()
df_competition
```

## find all unique competiton id + season id pair
Also print competition and season name to get better idea

We will proceed with champions league, laliga, worldcup and euro

```python
df_championsLeague_rows = df_competition[df_competition['competition_id'] == 16]
df_laliga_rows = df_competition[df_competition['competition_id'] == 11]
df_worldCup_rows = df_competition[df_competition['competition_id'] == 43]
df_euro_rows = df_competition[df_competition['competition_id'] == 55]
```

```python
df_championsLeague_rows
```

## Champions League contains only one match of a season

```python
import pandas as pd

df_championsLeague_matches = pd.DataFrame()

# Function to fetch matches based on 'competition_id' and 'season_id' values
def fetch_matches(row):
    competition_id = row['competition_id']
    season_id = row['season_id']
    # Your logic or function based on 'competition_id' and 'season_id' values
    df_match = parser.match(competition_id=competition_id, season_id=season_id)
    return df_match

# Iterate through the rows and fetch matches
for index, row in df_championsLeague_rows.iterrows():
    matches = fetch_matches(row)
    # Append the matches to df_championsLeague_matches
    df_championsLeague_matches = df_championsLeague_matches.append(matches, ignore_index=True)

# Print the appended DataFrame
print(df_championsLeague_matches)
```
